```python
BMAX = 20000
def bellcurve(x):
    mu = 0
    sig = 0.05*BMAX #knee curve starts at 25% of BMAX
    if(np.abs(x) <= 0.12*BMAX):
        norm_reward = (np.exp(-np.power((x - mu)/sig, 2.)/2) / np.exp(-np.power((0 - mu)/sig, 2.)/2))
    else:
        norm_reward = -0.25 -2.5*np.abs(x/BMAX)
    
    return norm_reward*400

def drewardfn(x):
        mu = 0
        sig = 1000
        if(np.abs(x) <= 2400): #24hr * 100mW/hr
            return ((1./(np.sqrt(2.*np.pi)*sig)*np.exp(-np.power((x - mu)/sig, 2.)/2)) * 1000000)
        else:
            return -100 - 0.05*np.abs(x)
```

```python
y1 = np.empty(1)
y2 = np.empty(1)

for x in np.arange(-BMAX/4,BMAX/4):
    y1 = np.append(y1,bellcurve(x))
    y2 = np.append(y2,drewardfn(x))
y1 = np.delete(y1, 0, 0) #remove the first row which is garbage
y2 = np.delete(y2, 0, 0) #remove the first row which is garbage


enp = np.arange(-BMAX/4,BMAX/4)
fig, ax = plt.subplots()
ax.plot(enp, y1)
ax.plot(enp, y2, 'r--')

ax.set(xlabel='ENP', ylabel='Reward',
       title='Reward Function')
ax.grid()
plt.show()
```

```python
BMAX = 20000
y1 = np.empty(1)
for x in np.arange(-BMAX/2,BMAX/2):
    y1 = np.append(y1,drewardfn(x))
y1 = np.delete(y1, 0, 0) #remove the first row which is garbage

enp = np.arange(-BMAX/2,BMAX/2)
fig, ax = plt.subplots()
ax.plot(enp, y1)

ax.set(xlabel='ENP', ylabel='Reward',
       title='Reward Function')
ax.grid()
plt.show()
```
